### 수질데이터 EDA

```python
water_quality_df = pd.read_csv(path + '환경 데이터/수질.csv')
```

```python
water_quality_df.head()
```

```python
water_quality_df['pH분류'].replace({'좋음':1, '나쁨':-1}, inplace=True)
water_quality_df['수온분류'].replace({'나쁨':1, '매우 나쁨':-1}, inplace=True)
water_quality_df['전도분류'].replace({'좋음':1, '나쁨':-1}, inplace=True)
water_quality_df.head()
```

```python
water_quality_df.columns
```

```python
df = pd.merge(df, water_quality_df, left_on='수질권역', right_on='측정소 위치', how='left')
df.head()
```

```python
df.drop(columns='측정소 위치', inplace=True)
df.info()
```

```python
df.head()
```

```python
df.columns
```

```python
temp = df.groupby('동')[['BOD', 'COD', 'TOC', 'SS', 'DO', 'T-P', '총대장균군', '분원성대장균군', '암모니아성질소', '질산성질소', '용존총질소', '인산염인', '용존총인', '클로로필A']].mean()
temp.head()
```

```python
df['pH분류'].isnull().sum()
```

```python
temp2 = df.groupby('동')[['pH분류', '수온분류', '전도분류']].min()
```

```python
temp2.head()
```

```python
temp3 = pd.concat([temp, temp2], axis=1)
temp3.info()
```

```python
temp3.fillna(0, inplace=True)
temp3.head()
```

```python
temp3.reset_index(inplace=True)
```

```python
drop_cols = ['BOD', 'COD', 'TOC', 'SS', 'DO', 'T-P', '총대장균군', '분원성대장균군', '암모니아성질소', '질산성질소', '용존총질소', '인산염인', '용존총인', '클로로필A', 'pH분류', '수온분류', '전도분류', 'pH', '전기전도도', '최저수온', '최고수온']
df.drop(columns=drop_cols, inplace=True)
df.head()
```

```python
df.info()
```

```python
df = pd.merge(df, temp3)
df.head()
```

```python
df.to_excel('통합데이터_수질.xlsx', index=False)
```
